## Carrier claims

```python
import pandas as pd
import numpy as np
import os
cwd = os.getcwd()

#If on MAC, this will likely work
datadir = '/'.join(cwd.split('/')[0:-2]) + '/Project/'
```

```python
#claim files for merging
claims1 = pd.read_csv(datadir + 'DE1_0_2008_to_2010_Carrier_Claims_Sample_2A.csv')
claims1.head()
```

```python
claims2 = pd.read_csv(datadir + 'DE1_0_2008_to_2010_Carrier_Claims_Sample_2B.csv')
claims2.head()
```

```python
#Merging two claims files
claims = pd.concat([claims1, claims2])
```

```python
working_df = claims
working_df.head()
```

```python
#Drop rows and columns where all elements are NULL
working_df = working_df.dropna(axis=1, how='all')
working_df = working_df.dropna(axis=0, how='all')
working_df.shape
```

```python
import pickle
clean_df = pickle.load(open(datadir +"working_df.p", "rb" ))
#clean_df = pd.read_csv('preprocessed_data.csv')
```

```python
clean_df.head()
```

```python
#Merge carrier claims with cleaned dataset
carrier = pd.merge(working_df, clean_df[['DESYNPUF_ID', 'end_train_dt', 'Y', 'TRUE_AGE']], how='left', on='DESYNPUF_ID')
```

```python
import datetime
from datetime import date
import dateutil
from datetime import datetime
from dateutil import parser
import numpy as np
import timeit

#Take a long time to run ~499 seconds, due to the size of the database
#If claim from date is NA, then drop
#Change claim from date to string like '20080207', and then to date format 2008-02-07

start = timeit.default_timer() 
carrier = carrier.dropna(0, subset = ['CLM_FROM_DT'])
carrier['CLM_FROM_DT'] = [str(int(x)) for x in carrier['CLM_FROM_DT']]
carrier['CLM_FROM_DT'] = [dateutil.parser.parse(str(date)).date() for date in carrier['CLM_FROM_DT']]
stop = timeit.default_timer() 
print (stop - start)
```

```python
#Check if any carrier claims are for patients who are not in cleaned dataset
carrier.end_train_dt.isnull().any()
```

```python
#Drop claims if there is no end_train_dt from clean dataset
carrier_1 = carrier.dropna(subset=['end_train_dt'])
carrier_1.head()
```

```python
print (carrier_1.shape)
```

```python
#Takes a long time to run, ~510 seconds
#Change end_train_date to a date format
start = timeit.default_timer() 
carrier_1['end_train_dt'] = [dateutil.parser.parse(str(date)).date() for date in carrier_1['end_train_dt']]
stop = timeit.default_timer() 
print (stop - start)
carrier_1.head()
```

```python
office_visits_pmt = pd.DataFrame(office_visits.groupby('DESYNPUF_ID').sum()['T_NCH_PMT'])
```
